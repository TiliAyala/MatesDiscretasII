El código a continuación genera la lista de números cuadrados más corta que, sumándose entre sí, dan con el número n ingresado por el usuario.


```python

import math

def Sumador(Target, NumCuadrado, Lista):
  if sum(Lista)<=Target: #Este condicional previene que el sumador añada números a la lista si esta ya superó o igualó el valor deseado.
    if (sum(Lista)+NumCuadrado)<= Target: #Este condicional previene que el valor que se va a añadir a la lista sobrepase al Target.
      Lista.append(NumCuadrado)
      Sumador(Target, NumCuadrado, Lista)



ListaElementos = []
print("Ingrese el valor al que se quiere llegar mediante suma de cuadrados: ")
NumeroTarget = int(input())

Contador = int(NumeroTarget-1)

while Contador>0:
  Root = math.sqrt(Contador)
  if Root%1 == 0:
    #Si el número pasó este filtro, es un número cuadrado.
    Sumador(NumeroTarget, Contador, ListaElementos)
  
  Contador = Contador - 1

print("La lista de números cuadrados más corta que suman el número deseado es: ")
print(ListaElementos)


```
