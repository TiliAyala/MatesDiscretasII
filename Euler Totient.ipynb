A continuación se presenta el cálculo de cualquier Euler Totient para cualquier número n, calculado de maneras diferentes. La primera manera utiliza fuerza bruta, descartando número por número hasta llegar a n. La segunda manera aprovecha las propiedades numéricas para incrementar la eficiencia del código.


```python
#MÉTODO DE FUERZA BRUTA

def mcd(a,b):
  if b == 0:
    return a
  return mcd(b, a%b)

ListaTotier = []

Valor = int(input("Introduzca el número al que se le va a sacar el Euler Totier: "))

n = 0

while n < Valor:
  resultado = mcd(Valor, n)
  if resultado == 1:
    ListaTotier.append(resultado)
  n = n+1

print("El Euler Totier de "+str(Valor)+" es "+str(len(ListaTotier)))


```

```python
#MÉTODO DE AHORRO DE RECURSOS

#La implementación de este método está basada en un lema que afirma:
#                  i. Euler Totier(1) = 1
#                  ii. Euler Totier (p^a) = p^a - p^(a-1) , siendo p un número primo.
#                  iii. Si mcd (m,n) = 1, entonces EulerTotier(m n)= EulerTotier(m) * EulerTotier(n)

def Descomposicion (n):

  NumerosPrimos = []

  for i in range(2,n+1):
    while n % i == 0:
      NumerosPrimos.append(i)
      n = n/i
  return NumerosPrimos

def TraduccionAPotencias (set, lista):

  ListaFinal = [] #Esta lista guarda en duplas los números primos, y a su derecha el valor de la potencia

  i = 0
  while i < len(set):
    target = list(set)
    potencia = lista.count(target[i])
    ListaFinal.append(target[i])
    ListaFinal.append(potencia)
    i = i+1
  #print(ListaFinal)
  return ListaFinal



def EulerTotierPrimo (lista):
  #Esta función únicamente aplica para números primos.
  i = 0

  EulerTotierFinal = 1

  while i< len(lista):
    EulerTotierFinal = EulerTotierFinal * (((lista[i])**(lista[i+1]))-(lista[i])**((lista[i+1])-1))
    i = i+2
  return EulerTotierFinal

# ------------------------------------------------------------------------- #



Valor = int(input("Introduzca el número al que se le va a sacar el Euler Totier: "))

DescomposicionEnPrimos = Descomposicion(Valor)
PrimosSinRepetir = set(DescomposicionEnPrimos)

PrimosFinal = TraduccionAPotencias(PrimosSinRepetir, DescomposicionEnPrimos)


EulerTotier = EulerTotierPrimo(PrimosFinal)

#print((DescomposicionEnPrimos))

print("El Euler Totier de "+str(Valor)+" es "+str(EulerTotier))



    
```
